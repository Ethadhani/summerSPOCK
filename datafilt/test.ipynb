```python
import rebound
import ploter
```

```python
sim = rebound.Simulation()
sim.add(m=1.)
sim.add(m=1.e-5, P=2, e=0, pomega = 0, l=0)
sim.add(m=0, P=3.6, e=0, pomega = 0, l=0)
sim.add(m=1.e-5, P=4, e=0, pomega = 0, l=0)


rebound.OrbitPlot(sim)
```

```python
ploter.insimPlot(sim,10000,20000)
```
